```python
def theoretical(a, b, alpha):
    (a, b) = (b - a, a) # In my analysis the damage = a * u + b, where u ~ U(0, 1)
    return ((alpha - (1/3) * alpha ** 2) * a + (2 * alpha - alpha ** 2) * b) / (alpha * ((1/2) * a + b)) - 1
```

```python
import random
def simulated(a, b, alpha, N=100000):
    
    without_potion = sum(random.uniform(a, b) if random.random() < alpha else 0 for _ in range(N)) / N
    with_potion = sum(max(
      (random.uniform(a, b) if random.random() < alpha else 0),
      (random.uniform(a, b) if random.random() < alpha else 0)
    ) for _ in range(N)) / N
    return (with_potion - without_potion) / without_potion
```

```python
print(f"{theoretical(0, 100, 1):.5f}, {simulated(0, 100, 1):.5f}")
print(f"{theoretical(50, 100, 1):.5f}, {simulated(50, 100, 1):.5f}")
print(f"{theoretical(1, 100, 0.7):.5f}, {simulated(1, 100, 0.7):.5f}")
print(f"{theoretical(100, 100, 0.7):.5f}, {simulated(100, 100, 0.7):.5f}")
print(f"{theoretical(20, 100, 0.2):.5f}, {simulated(20, 100, 0.2):.5f}")
```
